


User Login Check


```python
username = "admin"
password = "1234"

input_username = input("Enter username: ")
input_password = input("Enter password: ")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
def login_system(stored_username, stored_password):
    print("Login System")

    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == stored_username and password == stored_password:
        print("Login Successful\n")
        return True
    else:
        print("Invalid Credentials\n")
        return False
if __name__ == "__main__":

    # Login
    login_system("admin", "1234")
```

Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
def analyze_results(marks):
    print("Student Result Analyzer")

    pass_students = [m for m in marks if m >= 50]
    fail_students = [m for m in marks if m < 50]

    print(f"Total Students : {len(marks)}")
    print(f"Passed         : {len(pass_students)}")
    print(f"Failed         : {len(fail_students)}\n")
if __name__ == "__main__":
   marks = [45, 78, 90, 33, 60]
   analyze_results(marks)
```

Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters.\n")
```

Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print("Total ERROR messages:", error_count)
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = sum(1 for log in logs if log == "ERROR")
print("Total ERROR messages:", error_count)
```
