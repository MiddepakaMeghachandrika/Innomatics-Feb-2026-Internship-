```python
def employee_bonus_eligibility(employees: dict):
    """
    Identifies top-performing employees.
    Handles ties in highest performance score.
    """
    print("\n Employee Bonus Eligibility ")

    highest_score = max(employees.values())

    # Handle ties
    top_performers = [name for name, score in employees.items() if score == highest_score]

    print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
    return top_performers
if __name__ == "__main__":

    # Problem 1
    employees = {
        "Ravi": 92,
        "Anita": 88,
        "Kiran": 92,
        "Suresh": 85
    }
    employee_bonus_eligibility(employees)
```

```python
def keyword_analysis(query: str):
    """
    Converts query to lowercase,
    removes punctuation,
    counts word frequency,
    displays keywords appearing more than once.
    """
    print("\nSearch Query Keyword Analysis")

    import string

    # Normalize text
    query = query.lower()
    query = query.translate(str.maketrans("", "", string.punctuation))

    words = query.split()

    frequency = {}

    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    # Filter words appearing more than once
    repeated_keywords = {k: v for k, v in frequency.items() if v > 1}

    print(repeated_keywords)
    return repeated_keywords

if __name__ == "__main__":
   keyword_analysis("Buy mobile phone buy phone online")
```

```python
def sensor_validation(sensor_readings: list):
    """
    Identifies even-numbered readings as valid.
    Stores them as (hour_index, reading_value) pairs.
    """
    print("\n Sensor Data Validatio")

    valid_readings = [(index, value)
                      for index, value in enumerate(sensor_readings)
                      if value % 2 == 0]

    print("Valid Sensor Readings (Hour, Value):")
    print(valid_readings)
    return valid_readings
if __name__ == "__main__":
   sensor_readings = [3, 4, 7, 8, 10, 12, 5]
   sensor_validation(sensor_readings)
```

```python
def email_domain_analysis(emails: list):
    """
    Counts domain frequency and calculates percentage usage.
    """
    print("\n Email Domain Usage Analysis ")

    domain_count = {}

    for email in emails:
        domain = email.split("@")[1]
        domain_count[domain] = domain_count.get(domain, 0) + 1

    total = len(emails)

    for domain, count in domain_count.items():
        percentage = (count / total) * 100
        print(f"{domain}: {percentage:.0f}%")

    return domain_count
if __name__ == "__main__":
  emails = [
        "ravi@gmail.com",
        "anita@yahoo.com",
        "kiran@gmail.com",
        "suresh@gmail.com",
        "meena@yahoo.com"
    ]
  email_domain_analysis(emails)
```

```python
def sales_spike_detection(sales: list):
    """
    Calculates average sales.
    Detects days where sales are more than 30% above average.
    """
    print("\n Sales Spike Detection ")

    average_sales = sum(sales) / len(sales)
    threshold = average_sales * 1.30

    for day, value in enumerate(sales, start=1):
        if value > threshold:
            print(f"Day {day}: {value}")

    return average_sales
if __name__ == "__main__":
   sales = [1200, 1500, 900, 2200, 1400, 3000]
   sales_spike_detection(sales)

```

```python
def duplicate_user_detection(user_ids: list):
    """
    Identifies duplicate user IDs
    and displays their frequency.
    """
    print("\n Duplicate User ID Detection ")

    frequency = {}

    for user in user_ids:
        frequency[user] = frequency.get(user, 0) + 1

    duplicates = {user: count for user, count in frequency.items() if count > 1}

    for user, count in duplicates.items():
        print(f"{user} → {count} times")

    return duplicates
if __name__ == "__main__":
   user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
   duplicate_user_detection(user_ids)
```
