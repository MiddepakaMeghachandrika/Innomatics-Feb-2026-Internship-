```python
def post_engagement_analyzer():
    n = int(input("Enter number of posts: "))
    likes_list = []

    for i in range(n):
        likes = int(input(f"Enter likes for post {i+1}: "))
        likes_list.append(likes)

    total_likes = 0
    for like in likes_list:
        total_likes += like

    print(f"Total Likes: {total_likes}")

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


post_engagement_analyzer()
```

```python
def rainfall_checker():
    while True:
        try:
            n = int(input("Enter number of months (1-12): "))
            if 1 <= n <= 12:
                break
            else:
                print("Enter value between 1 and 12.")
        except:
            print(" Invalid input.")

    rainfall = []
    for i in range(n):
        while True:
            try:
                r = float(input(f"Enter rainfall for month {i+1} (0-500 mm): "))
                if 0 <= r <= 500:
                    rainfall.append(r)
                    break
                else:
                    print("Enter value between 0 and 500.")
            except:
                print("Invalid input.")

    required = 70
    avg = sum(rainfall) / n

    print("\nAverage Rainfall:", round(avg, 2))
    if avg >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rainfall_checker()
```

```python
def duplicate_accounts():
    while True:
        try:
            n = int(input("Enter number of usernames (1-100): "))
            if 1 <= n <= 100:
                break
            else:
                print("Enter between 1 and 100.")
        except:
            print("Invalid input.")

    users = []
    for i in range(n):
        name = input(f"Enter username {i+1}: ")
        users.append(name)

    if len(users) != len(set(users)):
        print("\nDuplicate Accounts Found: Yes")
    else:
        print("\nDuplicate Accounts Found: No")


duplicate_accounts()
```

```python
def appointment_check():
    while True:
        try:
            age = int(input("Enter patient age (0-120): "))
            if 0 <= age <= 120:
                break
            else:
                print("Enter valid age between 0 and 120.")
        except:
            print("Invalid input.")

    print("\nPatient Age:", age)
    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")


appointment_check()
```

```python
def premium_crop():
    while True:
        try:
            n = int(input("Enter number of crop prices (1-50): "))
            if 1 <= n <= 50:
                break
            else:
                print("Enter between 1 and 50.")
        except:
            print("Invalid input.")

    prices = []
    for i in range(n):
        while True:
            try:
                price = int(input(f"Enter crop price {i+1} (0-10000): "))
                if 0 <= price <= 10000:
                    prices.append(price)
                    break
                else:
                    print("Enter between 0 and 10000.")
            except:
                print("Invalid input.")

    premium = [p for p in prices if p > 2000]

    print("\nPremium Crops:", premium)


premium_crop()
```

```python
def app_health():
    while True:
        try:
            errors = int(input("Enter error count (0-100): "))
            if 0 <= errors <= 100:
                break
            else:
                print("Enter between 0 and 100.")
        except:
            print("Invalid input.")

    print("\nError Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


app_health()
```

```python
def transaction_check():
    while True:
        try:
            amount = int(input("Enter transaction amount (0-1000000): "))
            if 0 <= amount <= 1000000:
                break
            else:
                print("Enter between 0 and 1000000.")
        except:
            print(" Invalid input.")

    limit = 50000

    print("\nTransaction Amount:", amount)

    if amount <= limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


transaction_check()
```

```python
def attendance_check():
    while True:
        try:
            total = int(input("Enter total classes (1-500): "))
            attended = int(input("Enter attended classes (0-total): "))
            if 1 <= total <= 500 and 0 <= attended <= total:
                break
            else:
                print("Invalid values.")
        except:
            print("Invalid input.")

    percentage = (attended / total) * 100

    print("\nAttendance Percentage:", round(percentage, 2))

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance_check()
```

```python
def electricity_bill():
    while True:
        try:
            units = int(input("Enter electricity units (0-10000): "))
            if 0 <= units <= 10000:
                break
            else:
                print("Enter between 0 and 10000.")
        except:
            print(" Invalid input.")

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("\nTotal Bill: ₹", bill)
    print("Usage Status:", status)


electricity_bill()
```
