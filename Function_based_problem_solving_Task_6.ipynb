Smart Parking Lot Management System

```python
def smart_parking():
    # Capacity Input
    while True:
        try:
            capacity = int(input("Enter Parking Capacity (1-1000): "))
            if 1 <= capacity <= 1000:
                break
            else:
                print("Capacity must be between 1 and 1000.")
        except:
            print("Invalid input.")

    # Number of logs
    while True:
        try:
            n = int(input("Enter number of vehicle logs (1-1000): "))
            if 1 <= n <= 1000:
                break
            else:
                print("Enter between 1 and 1000.")
        except:
            print("Invalid input.")

    parked = 0
    peak = 0

    for i in range(n):
        while True:
            log = input(f"Enter log {i+1} (IN/OUT): ").upper()
            if log in ["IN", "OUT"]:
                break
            else:
                print("Enter only IN or OUT.")

        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    print("\nCurrently Parked Vehicles:", parked)
    print("Peak Usage:", peak)

    if parked > capacity:
        print("Parking Status: Exceeded Capacity ⚠")
    else:
        print("Parking Status: Available")


smart_parking()
```

Online Food Delivery Time Estimator

```python
def delivery_estimator():
    while True:
        try:
            distance = float(input("Enter Distance (1-50 km): "))
            if 1 <= distance <= 50:
                break
            else:
                print("Enter between 1 and 50 km.")
        except:
            print(" Invalid input.")

    traffic = input("Enter Traffic Level (Low/Medium/High): ").capitalize()
    weather = input("Enter Weather (Clear/Rainy/Storm): ").capitalize()

    eta = distance * 5  # base 5 mins per km

    # Traffic delay
    if traffic == "Medium":
        eta += 10
    elif traffic == "High":
        eta += 20

    # Weather delay
    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("\nEstimated Delivery Time:", int(eta), "minutes")


delivery_estimator()
```

Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy():
    while True:
        try:
            total = int(input("Enter Total Seats (1-1000): "))
            if 1 <= total <= 1000:
                break
            else:
                print(" Enter between 1 and 1000.")
        except:
            print("Invalid input.")

    while True:
        try:
            booked = int(input("Enter Number of Booked Seats (0-total): "))
            if 0 <= booked <= total:
                break
            else:
                print("Invalid booked seats.")
        except:
            print(" Invalid input.")

    occupancy = (booked / total) * 100

    print("\nOccupancy:", round(occupancy, 2), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")


theatre_occupancy()
```

Cloud Server Load Classification

```python
def server_load():
    while True:
        try:
            n = int(input("Enter number of CPU readings (1-100): "))
            if 1 <= n <= 100:
                break
            else:
                print("Enter between 1 and 100.")
        except:
            print("Invalid input.")

    readings = []
    for i in range(n):
        while True:
            try:
                cpu = float(input(f"Enter CPU reading {i+1} (0-100%): "))
                if 0 <= cpu <= 100:
                    readings.append(cpu)
                    break
                else:
                    print("Enter between 0 and 100.")
            except:
                print("Invalid input.")

    avg = sum(readings) / n

    print("\nAverage CPU Load:", round(avg, 2), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical ⚠")


server_load()
```

Smart Classroom Resource Monitor

```python
def classroom_monitor():
    resources = {}
    items = ["Projector", "AC", "Lights"]

    for item in items:
        while True:
            try:
                hours = int(input(f"Enter usage hours for {item} (0-24): "))
                if 0 <= hours <= 24:
                    resources[item] = hours
                    break
                else:
                    print(" Enter between 0 and 24.")
            except:
                print(" Invalid input.")

    overused = [k for k, v in resources.items() if v > 8]

    if overused:
        print("\nOverused Resources:", ", ".join(overused))
        print("Energy Alert: Yes ")
    else:
        print("\nOverused Resources: None")
        print("Energy Alert: No")


classroom_monitor()
```

Online Event Registration Capacity Controller

```python
def event_registration():
    while True:
        try:
            capacity = int(input("Enter Event Capacity (1-10000): "))
            if 1 <= capacity <= 10000:
                break
            else:
                print("Invalid capacity.")
        except:
            print(" Invalid input.")

    while True:
        try:
            registrations = int(input("Enter Total Registrations (0-20000): "))
            if 0 <= registrations <= 20000:
                break
            else:
                print(" Invalid registrations.")
        except:
            print(" Invalid input.")

    if registrations <= capacity:
        print("\nConfirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        print("\nConfirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")


event_registration()
```
