```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def switch(self, action):
        action = action.upper()
        if action in ["ON", "OFF"]:
            self.status = action
        else:
            print(" Invalid action. Use ON or OFF.")

    def display_status(self):
        print(f"{self.name} is {self.status}")


# User Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)
light.switch(action)
light.display_status()
```

```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# User Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = EmployeeIDCard(name, emp_id, dept)
emp.display_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# User Input
contact_name = input("Contact Name: ")
while True:
    try:
        phone = int(input("Phone Number: "))
        break
    except:
        print(" Enter valid number.")

contact = Contact(contact_name, phone)
contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))


# User Input
product_name = input("Product Name: ")
while True:
    try:
        price = float(input("Price: "))
        break
    except:
        print(" Enter valid price.")

product = Product(product_name, price)
product.display_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")


# User Input
movie_name = input("Movie Name: ")
while True:
    try:
        rating = float(input("Rating (0-5): "))
        if 0 <= rating <= 5:
            break
        else:
            print(" Enter rating between 0 and 5.")
    except:
        print(" Invalid input.")

movie = Movie(movie_name, rating)
movie.display_movie()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_delivery(self):
        print("\nDelivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# User Input
customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer_name, address)
delivery.display_delivery()
```
