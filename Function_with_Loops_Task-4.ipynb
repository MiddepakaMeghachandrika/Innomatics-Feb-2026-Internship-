```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print(" Recharge amount must be at least ₹50.")
        elif amount not in valid_plans:
            print(" Invalid plan. Available plans:", valid_plans)
        else:
            print(" Recharge Successful!")
            break   # exit loop if valid

        retry = input("Do you want to retry? (yes/no): ").lower()
        if retry != "yes":
            print("Transaction Cancelled.")
            break
if __name__ == "__main__":
    validate_recharge()
```

```python
def inventory_alert(products):
    print("\n--- Inventory Status ---")

    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")
if __name__ == "__main__":
  products = {
        "Laptop": 20,
        "Mouse": 10,
        "Keyboard": 8,
        "Monitor": 25
    }
  inventory_alert(products)
```

```python
def student_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"\nAverage Marks: {average}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
if __name__ == "__main__":
  result = student_result([60, 55, 70])
  print("Result:", result)
```

```python
def calculate_fare(distance, peak=False):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak:
        fare += fare * 0.25   # 25% extra

    return fare


def cab_system():
    while True:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        peak = True if peak_input == "yes" else False

        fare = calculate_fare(distance, peak)

        print(f"Estimated Fare: ₹{fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
if __name__ == "__main__":
  cab_system()
```

```python
def attendance_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print(f"\nAttendance Percentage: {percentage}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
if __name__ == "__main__":
  status = attendance_eligibility(["P", "P", "A", "P", "P"])
  print("Eligibility:", status)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password  (Must contain digit and special character)"
if __name__ == "__main__":
  print(check_password_strength("Pass@123"))
```
